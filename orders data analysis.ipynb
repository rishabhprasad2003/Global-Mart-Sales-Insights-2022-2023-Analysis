```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("ankitbansal06/retail-orders")

print("Path to dataset files:", path)
```

```python
import os

# Print all files in the dataset directory
for file in os.listdir(path):
    print(file)
```

```python
import pandas as pd

# Replace 'data.csv' with the actual file name
file_path = os.path.join(path, "orders.csv")
df = pd.read_csv(file_path)

# Display the first few rows
df.head()
```

```python
df.shape
```

##### Handling Null values

```python
df['Ship Mode'].unique()
```

As we see that values such as 'Not Available', 'unknown', 'nan' are all same 

```python
df = pd.read_csv(file_path, na_values=['Not Available', 'unknown'])
df.head()
```

```python
df['Ship Mode'].unique()
```

##### Now we will fix the column names for the dataset, making all the column name as lowercase and replacing ' ' with underscore '_'

```python
df.columns
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns = df.columns.str.replace(' ', '_')
```

```python
df.head()
```

##### Now making new columns as a feature engineering technique

```python
df['discount'] = df['cost_price']*df['discount_percent']*0.01
```

```python
df['sale_price'] = df['list_price']-df['discount']
```

```python
df['profit'] = df['sale_price']-df['cost_price']
```

```python
df.head()
```

##### Converting the 'order_date' column to date time format

```python
df['order_date'] = pd.to_datetime(df['order_date'], format="%Y-%m-%d")
```

```python
df.head()
```

```python
df.info()
```

##### Dropping the unnecesary columns

```python
df.drop(columns=['cost_price', 'list_price', 'discount_percent'], inplace=True)
```

when you try to query SELECT * FROM Orders;
PostgreSQL will look for a table named orders (in lowercase), not Orders.

1. To fix this:
If you want to access the table exactly as 'Orders' (with the capital "O"), you need to reference it in quotes, like this:

SELECT * FROM "Orders";

```python
# Load data into PostgreSQL
df.to_sql('Orders', con=engine, if_exists='replace', index=False)

print("Data loaded successfully!")
```

##### Now after creating and loading the table in postgresql
We will notice that the data types of the column are inappropriate
i.e., instead of `int` it is `bigint` which lead to more memory space.

As when creating tables in pandas it will take the data type of the highest value

##### Now that we know the problem, we will drop the table and create the table by giving the column our desired date type

And then append the values from the pandas dataframe to the sql database table

```python
df.to_sql('orders', con=engine, if_exists='append', index=False)

print("Data loaded successfully!")
```
